```python
##Primera funcion ---> numero mayor de tres
def maximo_de_tres(n1, n2, n3): 
    """Esta función toma tres números como argumentos y devuelve el mayor.""" 
    return max(n1, n2, n3)

print(maximo_de_tres(100, 50, 30))




```
